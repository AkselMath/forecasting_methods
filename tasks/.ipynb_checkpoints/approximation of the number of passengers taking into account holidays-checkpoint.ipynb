```python
import pandas as pd
import numpy as np
from scipy.optimize import minimize

import math

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
ausair = pd.read_csv('data/passengers.csv')
ausair = ausair['num_passengers']
```

```python
plt.figure(figsize=(12, 4), dpi=150)
plt.plot(ausair)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (12,5)
plot_acf(ausair, lags=48)
plt.xticks(np.arange(0, 50, 2))
plt.show()
```

```python
a = lambda y: pow(2, y)
np.array( list( map(a, range(3)) ) )
```

```python
def approximation(data: np.array, theta: np.array, beta: np.array) -> np.array:
    forcast = []
    for i in range(len(data)):
        a = lambda y: pow(i, y)
        pred = np.array( list( map(a, range(3)) ) ) @ theta + beta[0] * math.sin((2 * math.pi * (i-1)) / 12) + beta[1] * math.cos((2 * math.pi * (i-1)) / 12)
        forcast.append(pred)
    return np.array(forcast)
```

```python
def mean_square_error(data: np.array, theta: np.array, beta: np.array) -> int :
    forcast = approximation(data, theta, beta)
    return ((forcast - data)**2).mean()
```

```python
mse = lambda x: mean_square_error(ausair.values, x[0:3], x[3:])
    
df = minimize(mse, np.zeros(5))

df.x

theta = df.x[0: 3]
bata = df.x[3:]
```

```python
forcast = approximation(ausair.values, theta, bata)
mean_square_error(ausair.values, theta, bata)
```

```python
plt.figure(figsize=(12, 4), dpi=150)
plt.plot(ausair.values)
plt.plot(forcast)
plt.show()
```

```python
months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 
                  'August', 'September', 'October', 'November', 'December'] * 12)
months = pd.DataFrame(months, columns = ["months"])
months = pd.get_dummies(months['months'])
months['num_passengers'] = ausair.values
months.head(3)
```

```python
def approximation(data: np.array, theta: np.array, h: int) -> np.array:
    
    months = np.array(['January', 'February', 'March', 'April', 'May', 'June', 'July', 
                  'August', 'September', 'October', 'November', 'December'] * 12 * (h + 1))
    months = pd.DataFrame(months, columns = ["months"])
    months = pd.get_dummies(months['months'])
    
    forcast = []
    for i in range(len(data) + 12 * h):
        a = lambda y: pow(i, y)
        pred = np.array(list(map(a, range(3)))) @ theta[:3] + months.iloc[i].values @ theta[3:]
        forcast.append(pred)
    return np.array(forcast)
```

```python
def mean_square_error(data: np.array, theta: np.array) -> int :
    forcast = approximation(data, theta, 0)
    return ((forcast[:len(data)] - data)**2).mean()
```

```python
mse = lambda x: mean_square_error(ausair.values, x)

df = minimize(mse, np.zeros(15))
df.x

theta = df.x
```

```python
forcast = approximation(ausair.values, theta, 3)
mean_square_error(ausair.values, theta)
```

```python
plt.figure(figsize=(12, 4), dpi=150)
plt.plot(ausair.values)
plt.plot(forcast)
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf
```
